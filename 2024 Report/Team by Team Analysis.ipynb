```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('FF_2024.xlsx', sheet_name='positional_raw')
```

```python
df.head()
```

#### Get rank by position and week 

```python
df['Type'] = df['Group'].apply(lambda x: 'RB' if 'RB' in x else ('WR' if 'WR' in x else ('TE' if 'TE' in x else x))) 
df['rank_per_week'] = df.groupby(['Year ', 'Type', 'Week', 'Team'])['Points'].rank(ascending=False, method='min')
```

```python
df.head(22)
```

### By Position 

```python
dfr = df[(df['Week']<=15) & (df['Year ']==2024)]
qb = dfr[dfr['Group']=='QB']
rb = dfr[dfr['Group']=='RB']
wr = dfr[dfr['Group']=='WR']
te = dfr[dfr['Group']=='TE']
dst = dfr[dfr['Group']=='DEF']
k = dfr[dfr['Group']=='K']


wr1 = dfr[(dfr['Type']=='WR')&(dfr['rank_per_week']==1)]
wr2 = dfr[(dfr['Type']=='WR')&(dfr['rank_per_week']==2)]
wr3 = dfr[(dfr['Type']=='WR')&(dfr['rank_per_week']==3)]
wr4 = dfr[(dfr['Type']=='WR')&(dfr['rank_per_week']==4)]

rb1 = dfr[(dfr['Type']=='RB')&(dfr['rank_per_week']==1)]
rb2 = dfr[(dfr['Type']=='RB')&(dfr['rank_per_week']==2)]
rb3 = dfr[(dfr['Type']=='RB')&(dfr['rank_per_week']==3)]

qbd = pd.DataFrame(qb.groupby(['Team','Owner']).mean()['Points'])
dstd = pd.DataFrame(dst.groupby(['Team','Owner']).mean()['Points'])
kd = pd.DataFrame(k.groupby(['Team','Owner']).mean()['Points'])
```

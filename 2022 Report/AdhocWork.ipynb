```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
F2022 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2022')[0:12].drop('Place',1)
F2021 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2021')[0:12].drop('Place',1)
F2020 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2020')[0:12]
F2019 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2019')[0:12]
F2018 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2018')[0:12]
F2017 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2017')[0:12]
F2016 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2016')[0:12]
F2015 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2015')[0:12]
F2014 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2014')[0:12]
F2013 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2013')[0:12]
F2012 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2012')[0:12]
```

```python
Total = F2022.append(F2021.append(F2020.append(F2019.append(F2018.append(F2017.append(F2016 \
             .append(F2015.append(F2014.append(F2013.append(F2012))))))))))

All = Total.copy().reset_index(drop=True)
```

```python
All
```

# Worst 4-week starts 

```python
first = All[['Team_Name', 'Owner', 'Year', 'Week 1', 'Week 2','Week 3', 'Week 4']]
first['total4'] = first['Week 1'] + first['Week 2'] + first['Week 3'] + first['Week 4']
first['avg4'] = first['total4']/4
```

```python
first.sort_values(by='total4').reset_index(drop=True)
```

### Relative to baseline 

```python
arr = np.zeros((11,2))
for year in range(2012, 2023):
    subdf = first[first['Year']==year].copy()
    avg = np.mean(subdf['total4'])
    arr[year-2012,0] = year
    arr[year-2012,1] = avg

yearDF = pd.DataFrame(arr, columns=['Year','Baseline'])
yearDF['Baseline PPG'] = yearDF['Baseline']/4
```

```python
firstB = first.merge(yearDF, on=['Year']) 
firstB['PPG vs. Baseline'] = firstB['avg4'] - firstB['Baseline PPG']
```

```python
firstB.sort_values(by='PPG vs. Baseline').reset_index(drop=True).to_excel('worst_starts.xlsx')
```

# Points Against 

```python
pa2022 = pd.read_excel('FF_2022.xlsx', sheetname = 'points against')
```

```python
for year in range(2012, 2023):
    subdf = pa2022[pa2022['Year']==year].copy()
    avg = np.mean(subdf['Points Against'])
    arr[year-2012,0] = year
    arr[year-2012,1] = avg

yearDF = pd.DataFrame(arr, columns=['Year','Baseline'])
yearDF['Baseline PPG'] = np.where(yearDF['Year']>2020, yearDF['Baseline']/15, yearDF['Baseline']/14)
```

```python
pa2022['PPG'] = np.where(pa2022['Year']>2020, pa2022['Points Against']/15, pa2022['Points Against']/14)
pa2022B = pa2022.merge(yearDF, on=['Year']) 
pa2022B['PPG vs. Baseline'] = pa2022B['PPG'] - pa2022B['Baseline PPG']
```

```python
pa2022B.sort_values(by='PPG vs. Baseline', ascending=False).reset_index(drop=True).to_excel('pointsAgainst.xlsx')
```

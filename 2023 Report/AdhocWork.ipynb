```python
first = All[['Team_Name', 'Owner', 'Year', 'Week 1', 'Week 2','Week 3', 'Week 4','Week 5','Week 6','Week 7','Week 8']]
first['total8'] = first['Week 1'] + first['Week 2'] + first['Week 3'] + first['Week 4'] \
                + first['Week 5'] + first['Week 6'] + first['Week 7'] + first['Week 8']
first['avg8'] = first['total8']/8
```

```python
first.sort_values(by='total8').reset_index(drop=True)
```

### Relative to baseline 

```python
arr = np.zeros((12,2))
for year in range(2012, 2024):
    subdf = first[first['Year']==year].copy()
    avg = np.mean(subdf['total8'])
    arr[year-2012,0] = year
    arr[year-2012,1] = avg

yearDF = pd.DataFrame(arr, columns=['Year','Baseline'])
yearDF['Baseline PPG'] = yearDF['Baseline']/8
```

```python
firstB = first.merge(yearDF, on=['Year']) 
firstB['PPG vs. Baseline'] = firstB['avg8'] - firstB['Baseline PPG']
```

```python
firstB.sort_values(by='PPG vs. Baseline').reset_index(drop=True)
```

```python
firstB.sort_values(by='PPG vs. Baseline').reset_index(drop=True).to_excel('worst_starts.xlsx')
```

```python
def h2h_generator(owner, min_year=2020):
    sub = h2h[(h2h['OwnerW'].str.contains(owner))|(h2h['OwnerL'].str.contains(owner))]
    sub = sub[sub['Year']>=min_year]
    sub['win'] = np.where(sub['OwnerW'].str.contains(owner),1,0)
    sub['loss'] = np.where(sub['OwnerL'].str.contains(owner),1,0)
    sub['opponent'] = np.where(sub['win']==1, sub['OwnerL'], sub['OwnerW'])
    final = sub.groupby('opponent').sum()[['win','loss']]
    print(final['win'].sum(), final['loss'].sum())
    return final
```

```python
playoffs = h2h[h2h['Playoff']==1].reset_index(drop=True)
relevant = playoffs[(playoffs['Semi']==1)|(playoffs['Final']==1)].reset_index(drop=True)
```

```python
titles = relevant.groupby('OwnerW').sum()[['Final']].rename(columns={'Final':'Titles'}).reset_index()
```

```python
wins = relevant.groupby('OwnerW').count()[['Playoff']].rename(columns={'Playoff':'Wins'}).reset_index()
losses = relevant.groupby('OwnerL').count()[['Playoff']].rename(columns={'Playoff':'Losses'}).reset_index()
titles = relevant.groupby('OwnerW').sum()[['Final']].rename(columns={'Final':'Titles'}).reset_index()

record = wins.merge(losses, left_on='OwnerW', right_on='OwnerL', how='outer')
record = record.merge(titles, left_on='OwnerW', right_on='OwnerW', how='outer')

record['Owner'] = np.where(record['Wins']>=0, record['OwnerW'], record['OwnerL'])
record['Wins'] = np.where(record['Wins']>=0, record['Wins'], 0)
record['Losses'] = np.where(record['Wins']>=0, record['Losses'], 0)
record['Titles'] = np.where(record['Titles']>=0, record['Titles'], 0)
record['Appearances'] = np.where(record['Titles']>=0, record['Titles']+record['Losses'], record['Losses'])
record = record[['Owner','Appearances', 'Wins', 'Losses','Titles']].sort_values(by='Appearances',ascending=False)
```

# Avg Margin of Victory 

```python
h2h.head()
```

```python
def margin_generator(owner, year=2023):
    sub = h2h[(h2h['OwnerW'].str.contains(owner))|(h2h['OwnerL'].str.contains(owner))]
    sub = sub[sub['Year']==year]
    sub['win'] = np.where(sub['OwnerW'].str.contains(owner),1,0)
    sub['loss'] = np.where(sub['OwnerL'].str.contains(owner),1,0)

    subw = sub[sub['win']==1]
    avg_marginW = np.average(subw['Diff'])
    
    subL = sub[sub['loss']==1]
    avg_marginL = np.average(subL['Diff'])
    
    avg_margin = np.average(sub['Diff'])
    
    return avg_marginW, avg_marginL, avg_margin, year, owner
```

```python
def build_margin_table(year):
    h2h_yr = h2h[h2h['Year']==year]
    owners = h2h_yr['OwnerW'].unique()[:12]

    records = pd.DataFrame(pd.Series({'Owner':'blank', 'Year':year,
        'MarginW':0, 'MarginL':0, 'MarginT':0})).T


    for o in owners:
        avgW = np.round(margin_generator(o, year)[0],2)
        avgL = np.round(margin_generator(o, year)[1],2)
        avgT = np.round(margin_generator(o, year)[2],2)
        df_add = pd.DataFrame(pd.Series({'Owner':o,'Year':year,
                                     'MarginW':avgW,'MarginL':avgL,'MarginT':avgT})).T
        records = records.append(df_add)
        records = records[records['Owner']!='blank'].reset_index(drop=True)
    return records[['Owner','Year','MarginW','MarginL','MarginT']]
```

```python
combined = build_margin_table(2023).append(build_margin_table(2022)).append(build_margin_table(2021))
combined.sort_values(by='MarginL').reset_index(drop=True)
```

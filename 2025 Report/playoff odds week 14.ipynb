```python
import pandas as pd
import numpy as np
```

```python
mf_total = 1429.32 #jay
pd_total = 1377.86 #cam
ke_total = 1358.12 #kevin
to_total = 1458.46 #todd

sl_total = 1345.4 #tommy
gw_total = 1382.44 #brenton
pn_total = 1424.22 #mike

SL=0
GW=0
PN=0

SL2=0
GW2=0
PN2=0

SL3=0
GW3=0
PN3=0

MF=0 
PD=0
KE=0
TO=0

PD2=0
PD3=0
PD4=0

MF2=0
MF3=0
TO2=0
KE2=0

KE3 = 0
MF4 = 0

for i in range(10000):
    #gw = np.random.normal(134.68, 10) #brenton
    gw = 148.64
    #sl = np.random.normal(117.81, 10) #tommy
    sl = 119.96
    pn = np.random.normal(97.64, 7) #mike

    
    pd = np.random.normal(98.36, 7) #cam
    to = np.random.normal(137.48, 10) #todd
    mf = np.random.normal(60.65, 4) #jay
    ke = 83.48
    
    
    #tommy vs. brenton
    if sl+sl_total < gw+gw_total:
        SL=SL+1
    elif gw+gw_total < sl+sl_total:
        GW=GW+1 
    
    #tommy vs. mike
    if sl+sl_total < pn+pn_total:
        SL2=SL2+1
    elif pn+pn_total < sl+sl_total:
        PN=PN+1 

    #brenton vs. mike
    if gw+gw_total < pn+pn_total:
        GW2=GW2+1
    elif pn+pn_total < gw+gw_total:
        PN2=PN2+1 

    #three-way
    if gw+gw_total < pn+pn_total and gw+gw_total < sl+sl_total :
        GW3=GW3+1
    if pn+pn_total < sl+sl_total and pn+pn_total < gw+gw_total :
        PN3=PN3+1
    else: SL3=SL3+1

    #todd vs. cam

    if to+to_total > pd+pd_total:
        TO=TO+1
    elif to+to_total < pd+pd_total:
        PD=PD+1 

    #kevin vs. cam

    if ke+ke_total > pd+pd_total:
        KE=KE+1
    elif ke+ke_total < pd+pd_total:
        PD2=PD2+1 

  #jay vs. cam

    if mf+mf_total > pd+pd_total:
        MF=MF+1
    elif mf+mf_total < pd+pd_total:
        PD3=PD3+1 

 #todd vs. jay

    if mf+mf_total > to+to_total:
        MF2=MF2+1
    elif mf+mf_total < to+to_total:
        TO2=TO2+1 

 #kevin vs. jay

    if mf+mf_total > ke+ke_total:
        MF3=MF3+1
    elif mf+mf_total < ke+ke_total:
        KE2=KE2+1 


   #three-way: Cam, Kevin, Jay
    if pd+pd_total < ke+ke_total and pd+pd_total < mf+mf_total :
        PD4=PD4+1
    if ke+ke_total < pd+pd_total and ke+ke_total < mf+mf_total :
        KE3=KE3+1
    else: MF4=MF4+1
```

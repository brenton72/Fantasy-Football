```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import warnings
warnings.filterwarnings('ignore')
```

```python
F2025 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2025')[0:12]
F2024 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2024')[0:12]
F2023 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2023')[0:12]
F2022 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2022')[0:12]
F2021 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2021')[0:12]
F2020 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2020')[0:12]
F2019 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2019')[0:12]
F2018 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2018')[0:12]
F2017 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2017')[0:12]
F2016 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2016')[0:12]
F2015 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2015')[0:12]
F2014 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2014')[0:12]
F2013 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2013')[0:12]
F2012 = pd.read_excel('../LeagueHistory.xlsx', sheet_name='2012')[0:12]
```

```python
Total = pd.concat([F2025, F2024, F2023, F2022, F2021, F2020, F2019, F2018, F2017, F2016, F2015, F2014, F2013, F2012])


AvgPerSeason = {'2012': 0, '2013': 0, '2014': 0, '2015': 0, '2016': 0, 
                '2017': 0, '2018': 0, '2019':0, '2020':0, '2021':0, '2022':0, '2023':0,
               '2024': 0, '2025': 0}

for year in range(2012, 2026):
    subdf = Total[Total['Year']==year]
    average = np.mean(subdf['TOTAL'])
    AvgPerSeason[str(year)] = average
    
PerYear = pd.DataFrame.from_dict(AvgPerSeason, orient='index').reset_index(). \
rename(columns={'index':'Year', 0:'Baseline'})

PerYear['Year'] = pd.to_numeric(PerYear['Year'])
PerYear['BaselinePPG'] = np.where(((PerYear['Year'] < 2021) | (PerYear['Year'] > 2024)), PerYear['Baseline']/14, PerYear['Baseline']/15)#alternative code for newer version of Python


Total2 = Total.merge(PerYear,on=['Year'])
Total2['Points Above League Average'] = Total2['TOTAL'] - Total2['Baseline']
Total2['PPG Above League Average'] = Total2['Mean'] - Total2['BaselinePPG']
Total2['Year'] = Total2['Year'].astype(int)

All = Total2.copy()

Total2 = Total2.reset_index(drop=True)[['Team_Name','Owner', 'Year', 'TOTAL','Mean','Std','Playoffs', 'Place',
                                       'Points Above League Average', 'PPG Above League Average']]
PPR = Total2[(Total2['Year']>=2021)]
```

# Worst 7-game starts

```python
def first7(df):
    df2 = df[['Team_Name', 'Owner', 'Year', 'Week 1', 'Week 2','Week 3','Week 4','Week 5','Week 6','Week 7']]
    df2['total'] = 0
    for i in range(1, 8):
        df2['total'] = df2['total'] + df2['Week ' + str(i)]
    df2['PPG'] = df2['total']/7
    df3 = df2[['Team_Name','Owner','Year','PPG']].sort_values(by='PPG').reset_index(drop=True)

    AvgPerSeason7 = {'2012': 0, '2013': 0, '2014': 0, '2015': 0, '2016': 0, 
                '2017': 0, '2018': 0, '2019':0, '2020':0, '2021':0, '2022':0, '2023':0,
               '2024': 0}

    for year in range(2012, 2025):
        subdf = df3[df3['Year']==year]
        average = np.mean(subdf['PPG'])
        AvgPerSeason7[str(year)] = average

    PerYear = pd.DataFrame.from_dict(AvgPerSeason7, orient='index').reset_index(). \
            rename(columns={'index':'Year', 0:'Baseline7'})

    PerYear['Year'] = pd.to_numeric(PerYear['Year'])

    df4 = df3.merge(PerYear,on=['Year'])
    df4['baseline_diff'] = df4.PPG - df4.Baseline7
    df4 = df4.sort_values(by='baseline_diff').reset_index(drop=True)

    return df4
```

# First seven games vs. Last seven games

```python
All
```

```python
def first7(df):
    df2 = df[['Team_Name', 'Owner', 'Year', 'Week 1', 'Week 2','Week 3','Week 4','Week 5','Week 6','Week 7']]
    df2['total'] = 0
    for i in range(1, 8):
        df2['total'] = df2['total'] + df2['Week ' + str(i)]
    df2['PPG'] = df2['total']/7
    df3 = df2[['Team_Name','Owner','Year','PPG']].sort_values(by='PPG').reset_index(drop=True)

    AvgPerSeason7 = {'2012': 0, '2013': 0, '2014': 0, '2015': 0, '2016': 0, 
                '2017': 0, '2018': 0, '2019':0, '2020':0, '2021':0, '2022':0, '2023':0,
               '2024': 0}

    for year in range(2012, 2025):
        subdf = df3[df3['Year']==year]
        average = np.mean(subdf['PPG'])
        AvgPerSeason7[str(year)] = average

    PerYear = pd.DataFrame.from_dict(AvgPerSeason7, orient='index').reset_index(). \
            rename(columns={'index':'Year', 0:'Baseline7'})

    PerYear['Year'] = pd.to_numeric(PerYear['Year'])

    df4 = df3.merge(PerYear,on=['Year'])
    df4['baseline_diff'] = df4.PPG - df4.Baseline7
    df4 = df4.sort_values(by='baseline_diff').reset_index(drop=True)

    return df4
```

```python
sub = All[['Team_Name', 'Owner', 'Year', 'Week 1', 'Week 2','Week 3','Week 4','Week 5','Week 6','Week 7',
          'Week 8', 'Week 9','Week 10','Week 11','Week 12','Week 13','Week 14']]
week_cols1 = [f"Week {i}" for i in range(1, 8)]
week_cols2 = [f"Week {i}" for i in range(8, 15)]


sub["avg_points_1st"] = sub[week_cols1].mean(axis=1)
sub["avg_points_2nd"] = sub[week_cols2].mean(axis=1)
sub["diff"] = sub["avg_points_2nd"] - sub["avg_points_1st"]
```

```python
results = sub[['Team_Name', 'Owner','Year', 'avg_points_1st','avg_points_2nd', 'diff']].sort_values(by='diff')
```

```python
results.to_excel('seven_games.xlsx')
```

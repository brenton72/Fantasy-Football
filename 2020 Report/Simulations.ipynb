```python
std = 22

def simulate_one_season_scores(df):
    Brady_scores = np.random.normal(df.loc[0, 'Mean'], std, 14)
    Wakandan_scores = np.random.normal(df.loc[1, 'Mean'], std, 14)
    MrFan_scores = np.random.normal(df.loc[2, 'Mean'], std, 14)
    Sack_scores = np.random.normal(df.loc[3, 'Mean'], std, 14)
    Schmop_scores = np.random.normal(df.loc[4, 'Mean'], std, 14)
    Herbin_scores = np.random.normal(df.loc[5, 'Mean'], std, 14)
    Shiva_scores = np.random.normal(df.loc[6, 'Mean'], std, 14)
    Gee_scores = np.random.normal(df.loc[7, 'Mean'], std, 14)
    Micro_scores = np.random.normal(df.loc[8, 'Mean'], std, 14)
    BigW_scores = np.random.normal(df.loc[9, 'Mean'], std, 14)
    Jasjaap_scores = np.random.normal(df.loc[10, 'Mean'], std, 14)
    GoBias_scores = np.random.normal(df.loc[11, 'Mean'], std, 14)
    
    SCORES = np.array([Brady_scores, Wakandan_scores, MrFan_scores, Sack_scores, 
                       Schmop_scores, Herbin_scores, Shiva_scores, Gee_scores, 
                       Micro_scores, BigW_scores, Jasjaap_scores, GoBias_scores])
    return(SCORES)
```

```python
def week_fair(scores_array, wins_array, week = 12):
    Teams = {'Wolf':0, 'Eli':1, 'Cam':2, 'Will':3, 'Brenton':4, 
             'Burnett':5, 'Jay':6, 'Mike':7, 'Tommy':8, 
             'Todd':9, 'Kevin':10, 'Jasjaap': 11} 
    
    p = np.arange(12)
    random.shuffle(p)
    
    for n in range(6):
        first = p[2*n]
        second = p[2*n+1]
    
        if scores_array[first][week-1] > scores_array[second][week-1]:
            wins_array[first] = wins_array[first] + 1
        else:
            wins_array[second] = wins_array[second] + 1
    
    return wins_array 
```

```python
def simulate_one_season_matchups(scoring_array, fair=False):
    wins_array = np.zeros(12)
    WINS1 = week1(scoring_array, wins_array.copy())
    WINS2 = week2(scoring_array, WINS1.copy())
    WINS3 = week3(scoring_array, WINS2.copy())
    WINS4 = week4(scoring_array, WINS3.copy())
    WINS5 = week5(scoring_array, WINS4.copy())
    WINS6 = week6(scoring_array, WINS5.copy())
    WINS7 = week7(scoring_array, WINS6.copy())
    WINS8 = week8(scoring_array, WINS7.copy())
    WINS9 = week9(scoring_array, WINS8.copy())
    WINS10 = week10(scoring_array, WINS9.copy())
    WINS11 = week11(scoring_array, WINS10.copy())
    
    if fair==False:
        WINS12 = week12(scoring_array, WINS11.copy())
        WINS13 = week13(scoring_array, WINS12.copy())
        wins_per_team = tuple(week14(scoring_array, WINS13.copy()))
    elif fair==True:
        WINS12 = week_fair(scoring_array, WINS11.copy(), week=12)
        WINS13 = week_fair(scoring_array, WINS12.copy(), week=13)
        wins_per_team = tuple(week_fair(scoring_array, WINS13.copy(), week=14))

    #return final results
    losses_per_team = tuple([14-w for w in wins_per_team])
    score_per_team = tuple(scoring_array.sum(axis=1))
    label_per_team =  tuple(FF_2020_summary['Team_Name'])
    standings = pd.DataFrame(np.vstack((wins_per_team, losses_per_team, score_per_team)).T, index=label_per_team)
    standings = standings.rename(columns={0:'W', 1:'L', 2:'Pts'})    
    
    standings['Pts'] = standings['Pts'].astype(float)
    standings = standings.sort_values(by=['W', 'Pts'], ascending=[False, False])
    
    standings['Place'] = np.arange(1,13)
    standings['Playoffs'] = np.where(standings['Place'] < 5, 1, 0)
    standings['Top'] = np.where(standings['Place'] == 1, 1, 0)
    standings['Sacko'] = np.where(standings['Place'] == 12, 1, 0)
    
    places = np.array(standings['Place'])
    playoffs = np.array(standings['Playoffs'])
    tops = np.array(standings['Top'])
    sackos = np.array(standings['Sacko'])
    
    return standings
```

```python
def simulate_many_seasons(FF_summary, sims = 2, fair=False):
    '''
    @sims is the number of simulations you would like to run.
    '''
    SCORES = simulate_one_season_scores(FF_summary)
    TOTAL = simulate_one_season_matchups(SCORES, fair=fair)
    
    for i in range(1,sims):
            
        SCORES = simulate_one_season_scores(FF_summary)
        SEASON_DF = simulate_one_season_matchups(SCORES, fair=fair)
        
        for col in TOTAL.columns.values:
            TOTAL[col] = TOTAL[col]/(i+1)*i + SEASON_DF[col]/(i+1) #Average is continusously updated across trials
    return TOTAL

def clean_total(total):
    total['W'] = round(total['W'], 2)
    total['L'] = 14-total['W']
    total['Pts'] = round(total['Pts'], 2)
    total['Place'] = round(total['Place'], 2)
    total['Playoffs'] = round(total['Playoffs'], 3)*100
    total['Top'] = round(total['Top'], 3)*100
    total['Sacko'] = round(total['Sacko'], 3)*100
    total = total.sort_values(by='Place')
    return total
```

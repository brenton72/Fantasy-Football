```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
s2017 = pd.read_excel('LeagueHistory.xlsx', sheetname='2017')
s2016 = pd.read_excel('LeagueHistory.xlsx', sheetname='2016')
s2015 = pd.read_excel('LeagueHistory.xlsx', sheetname='2015')
s2014 = pd.read_excel('LeagueHistory.xlsx', sheetname='2014')
s2013 = pd.read_excel('LeagueHistory.xlsx', sheetname='2013')
s2012 = pd.read_excel('LeagueHistory.xlsx', sheetname='2012')
```

```python
m2012 = np.mean(s2012['Mean'])
m2013 = np.mean(s2013['Mean'])
m2014 = np.mean(s2014['Mean'])
m2015 = np.mean(s2015['Mean'])
m2016 = np.mean(s2016['Mean'])
m2017 = np.mean(s2017['Mean'])

m2012t = np.mean(s2012['TOTAL'])
m2013t = np.mean(s2013['TOTAL'])
m2014t = np.mean(s2014['TOTAL'])
m2015t = np.mean(s2015['TOTAL'])
m2016t = np.mean(s2016['TOTAL'])
m2017t = np.mean(s2017['TOTAL'])

means = [m2012, m2013, m2014, m2015, m2016, m2017]
meansT = [m2012t, m2013t, m2014t, m2015t, m2016t, m2017t]

print(means)
print(meansT)
```

```python
s2012['PPG Margin'] = s2012['Mean'] - m2012
s2013['PPG Margin'] = s2013['Mean'] - m2013
s2014['PPG Margin'] = s2014['Mean'] - m2014
s2015['PPG Margin'] = s2015['Mean'] - m2015
s2016['PPG Margin'] = s2016['Mean'] - m2016
s2017['PPG Margin'] = s2017['Mean'] - m2017

s2012['Pts Margin'] = s2012['TOTAL'] - m2012t
s2013['Pts Margin'] = s2013['TOTAL'] - m2013t
s2014['Pts Margin'] = s2014['TOTAL'] - m2014t
s2015['Pts Margin'] = s2015['TOTAL'] - m2015t
s2016['Pts Margin'] = s2016['TOTAL'] - m2016t
s2017['Pts Margin'] = s2017['TOTAL'] - m2017t
```

```python
plt.plot(means, marker='o')
plt.xticks(np.arange(6), ['2012', '2013', '2014', '2015', '2016', '2017'])
plt.grid(axis='y', lw=0.4)
plt.ylim([92.5, 100.5])
plt.ylabel('Points Per Game Per Team', size=13)
plt.yticks(size=13)
plt.xticks(size=13)

plt.title('Leaguewide Decline in Scoring', size = 16)
plt.show()
```

```python
ALL = s2017.append(s2016.append(s2015.append(s2014.append(s2013.append(s2012)))))
ALL['Points Per Game'] = round(ALL['Mean'],2)
ALL['Points'] = round(ALL['TOTAL'],2)


OVERALL = ALL[['Team_Name', 'Owner', 'Year', 'Playoffs', 'Points', 'Points Per Game', 'Pts Margin', 'PPG Margin']]

```

```python
top10overall = OVERALL.sort_values(by='Points', ascending=False).reset_index(drop=True).head(36)
worst10overall = OVERALL.sort_values(by='Points', ascending=True).reset_index(drop=True).head(10)

#top10overall.to_excel('top10.xlsx')
```

```python
import pandas as pd
import numpy as np
```

```python
F2021 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2021')[0:12].drop('Place',1)
F2020 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2020')[0:12]
F2019 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2019')[0:12]
F2018 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2018')[0:12]
F2017 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2017')[0:12]
F2016 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2016')[0:12]
F2015 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2015')[0:12]
F2014 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2014')[0:12]
F2013 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2013')[0:12]
F2012 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2012')[0:12]
```

```python
F2021
```

```python
Total = F2021.append(F2020.append(F2019.append(F2018.append(F2017.append(F2016 \
             .append(F2015.append(F2014.append(F2013.append(F2012)))))))))

All = Total.copy()

Total = Total.reset_index(drop=True)[['Team_Name','Owner', 'Year', 'TOTAL','Mean','Std','Playoffs']]
Total
```

```python
AvgPerSeason = {'2012': 0, '2013': 0, '2014': 0, '2015': 0, '2016': 0, 
                '2017': 0, '2018': 0, '2019':0, '2020':0, '2021':0}

for year in range(2012, 2022):
    subdf = Total[Total['Year']==year]
    average = np.mean(subdf['TOTAL'])
    AvgPerSeason[str(year)] = average
    
PerYear = pd.DataFrame.from_dict(AvgPerSeason, orient='index').reset_index(). \
rename(columns={'index':'Year', 0:'Baseline'})

PerYear['Year'] = pd.to_numeric(PerYear['Year'])
PerYear['BaselinePPG'] = np.where(PerYear['Year'] < 2021, PerYear['Baseline']/14, PerYear['Baseline']/15)
```

```python
Total2 = Total.merge(PerYear,on=['Year'])
Total2['Points Above League Average'] = Total2['TOTAL'] - Total2['Baseline']
Total2['PPG Above League Average'] = Total2['Mean'] - Total2['BaselinePPG']
Total2['Year'] = Total2['Year'].astype(int)
Total2
```

### Best Teams 

```python
best = Total2.sort_values(by='Mean', 
            ascending=False)[['Team_Name','Owner','Year', 'TOTAL', 'Mean',
            'Points Above League Average', 'PPG Above League Average']].reset_index(drop=True)

best2 =Total2.sort_values(by='PPG Above League Average', 
            ascending=False)[['Team_Name','Owner','Year', 'TOTAL', 'Mean',
            'Points Above League Average', 'PPG Above League Average']].reset_index(drop=True)
```

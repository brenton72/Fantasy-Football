```python
import pandas as pd
import numpy as np
```

```python
pa = pd.read_excel('FF_2021.xlsx', sheetname = 'points_against')
```

## Unluckiest teams 

### Get baseline for each year 

```python
df = pd.DataFrame()
for y in np.unique(pa['Year']):
    sub_df = pa[pa['Year']==y]
    avg = np.average(sub_df['Points Against'])
    sub_df['Baseline'] = avg
    if y < 2021: 
        sub_df['PPGA'] = sub_df['Points Against']/14
        sub_df['Baseline Avg'] = sub_df['Baseline']/14
    else:
        sub_df['PPGA'] = sub_df['Points Against']/15
        sub_df['Baseline Avg'] = sub_df['Baseline']/15
    df = df.append(sub_df)

df['Diff'] = df['PPGA']-df['Baseline Avg']
df = df.reset_index(drop=True)
```

```python
df.sort_values(by='Diff', ascending=False).reset_index(drop=True)
```

```python
df2021 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2021')
df2020 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2020')
df2019 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2019')
df2018 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2018')
df2017 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2017')
df2016 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2016')
df2015 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2015')
df2014 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2014')
df2013 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2013')
df2012 = pd.read_excel('LeagueHistory.xlsx', sheet_name='2012')

DF = df2021.append(df2020.append(df2019.append(df2018.append(df2017.append(df2016.append(df2015.append(df2014.append(df2013.append(df2012))
                                                                                                      )))))))

DF = DF[DF['Year'].notnull()]
```

```python
DF
```

```python
DF['Year'] = DF['Year'].astype(int)

DF['1st_half_avg'] = np.where(DF['Year']<2021, (DF['Week 1']+DF['Week 2']+DF['Week 3']+
                         DF['Week 4']+DF['Week 5']+DF['Week 6']+DF['Week 7'])/7,
                         (DF['Week 1']+DF['Week 2']+DF['Week 3']+
                         DF['Week 4']+DF['Week 5']+DF['Week 6']+DF['Week 7']+DF['Week 8'])/8
                         )

DF['2nd_half_avg'] = np.where(DF['Year']<2021, (DF['Week 8']+DF['Week 9']+DF['Week 10']+
                         DF['Week 11']+DF['Week 12']+DF['Week 13']+DF['Week 14'])/7,
                         (DF['Week 9']+DF['Week 10']+DF['Week 11']+
                         DF['Week 12']+DF['Week 13']+DF['Week 14']+DF['Week 15'])/7
                         )

DF['2nd half improvement'] = DF['2nd_half_avg']-DF['1st_half_avg']
```

```python
DF[['Team_Name', 'Owner', 'Year', '1st_half_avg', '2nd_half_avg', '2nd half improvement']]. \
            sort_values(by='2nd half improvement',ascending=True).reset_index(drop=True)
```
